# Popularity of Sociology Concepts

## Packages

```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
plt.xkcd()
```

## Load data

```python
papers = pd.read_csv('data/soc_concept_counts_wos.tsv', sep='\t')
papers = papers.set_index('pubyear')
papers.head()
```

```python
cites = pd.read_csv('data/soc_concept_cited_counts_wos.tsv', sep='\t')
cites = cites.set_index('pubyear')

cites.head()
```

```python
## Terms
```

```python
terms = ['intersectional',  'bell curve', 
         'creative class', 'clash of civilizations',  
         'bowling alone', 'second shift',
         'culture of fear','overworked americans', 
         ]
```

## Plot papers using the terms

```python
fig, axs = plt.subplots(4,2, sharex=True, figsize=(14,14))

for i, t in enumerate(terms):
    axs[int(i/2)][i%2].set_title(t)
    axs[int(i/2)][i%2].plot(papers[t+' in title'], label='In Title')
    axs[int(i/2)][i%2].plot(papers[t+' in abstract'], label='In Abstract')
    axs[int(i/2)][i%2].legend(loc='upper left')
    axs[int(i/2)][i%2].set_xlim((1980, 2018))
    
axs[3][0].set_xlabel('Year')
axs[3][1].set_xlabel('Year')
for i in range(0,4):
    axs[i][0].set_ylabel('Count')
    
fig.suptitle('Papers in the Web of Science that Mention Key Terms')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig('papers.png', dpi=300)
```

## Plot papers citing the terms

```python
fig, axs = plt.subplots(4,2, sharex=True, figsize=(14,14))

for i, t in enumerate(terms):
    axs[int(i/2)][i%2].set_title(t)
    axs[int(i/2)][i%2].plot(cites[t+' in title'], label='Title')
    axs[int(i/2)][i%2].plot(cites[t+' in abstract'], label='Abstract')
    axs[int(i/2)][i%2].legend(loc='upper left')
    axs[int(i/2)][i%2].set_xlim((1980, 2018))
    
axs[3][0].set_xlabel('Year')
axs[3][1].set_xlabel('Year')
for i in range(0,4):
    axs[i][0].set_ylabel('Count')
    
fig.suptitle('Citations to Papers in the Web of Science that Mention Key Terms')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig('cites_to.png', dpi=300)
```
